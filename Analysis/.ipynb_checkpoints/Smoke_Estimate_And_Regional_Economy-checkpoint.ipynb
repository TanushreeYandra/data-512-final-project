More and more frequently, summers in the western US have been characterized by wildfires with smoke billowing across multiple western states. There is a growing body of work pointing to the negative impacts of smoke on health, tourism, property, and other aspects of society. This project analyzes wildfire impacts on the city of Twin Falls, Idaho in the US. The end goal is to be able to inform policy makers, city managers, city councils, or other civic institutions, to make an informed plan for how they could or whether they should make plans to mitigate future impacts from wildfires.

This analysis thus studies the wildfires in detail and is divided into two parts. The first part focuses on studying the wildland fires within 1250 miles of the city of Twin Falls, Idaho for the last 60 years (1963-2020). A smoke estimate is then created to estimate the wildfire smoke impact which is later modeled to make predictions for the next 30 years (until 2049). The second part of the project further extends this analysis to a specific impact focus - regional socio-economy and assesses how this sector is impacted by wildfires.

This section of the notebook focusses on the second part of the analysis where four socio-economic indicators are chosen to compare with the modeled smoke estimate from [Wildfires Analysis - Smoke Estimate Modeling notebook]() and see for any correlation. If correlation does exist, predictions are generated for those indicators to understand how the wildfires will impact the future economy.

### Research Questions and Data for the Study

Four indicators - Gross Domestic Product (GDP), unemployment rate, income, and income inequality form the basis for this study. The first indicator - GDP, a widely used indicator referring to the total gross value added by all industries in the economy, was chosen to understand the overall impact of wildfires on the region’s economy. Next, in order to choose socio-economic indicators, the article from [Centers for Disease Control and Prevention](https://www.cdc.gov/dhdsp/health_equity/socioeconomic.htm) was referred to. The top three socio-economic indicators were defined as education, employment, and income. However, since there was no education level data available for the city of Twin Falls, Idaho, only employment and income data were considered. Finally, to understand if job losses and decrease in income in turn aggravate economic disparity, income inequality was chosen as the fourth indicator.

Specifically, the analysis wanted to focus on four research questions,

1. Do the years affected by heavy smoke experience dips in economic productivity? The aim is to assess the impact on GDP growth during periods of intense wildfires. 

2. Do unemployment rates increase during and immediately after wildfire events due to disruptions in economic activities? The goal is to analyze how wildfires affect unemployment rates in periods experiencing severe smoke exposure.

3. Did years with higher smoke exposure experience declines in personal income? The goal is to investigate how wildfires influence personal income per capita.

4. Finally, do low-income groups suffer disproportionately due to wildfires’ economic impacts? The aim is to explore if income inequality exacerbates the economic vulnerability of low-income groups during and after wildfires.


To answer the above research questions, the four indicators were sourced from four different datasets from the [Federal Reserve Economic Data (FRED)](https://fredhelp.stlouisfed.org/fred/about/about-fred/what-is-fred/), maintained by the Research Department at the Federal Reserve Bank of St. Louis. It is an online database consisting of hundreds of thousands of economic data time series from multiple national, international, public, and private sources. All the four datasets are under the [FredⓇ Services General License](https://fred.stlouisfed.org/legal/#property-rights). They are under copyright; but, provided that one has not engaged in any prohibited uses, one may use these data series with proper attribution of the source. For more information, please refer to the [Terms of Use](https://fred.stlouisfed.org/legal/#fred-terms). The four data sources and their uses have been mentioned below:

1. Decrease in Economic Output: For this, the dataset of [Real Gross Domestic Product: All Industries in Twin Falls County](https://fred.stlouisfed.org/series/REALGDPALL16083) was used. This is an annual dataset from 2001 to 2021 representing the GDP - a measure of the market value of final goods and services produced within a county in a particular period. The dataset contains only two columns - date in the YYYY-MM-DD format, and the GDP.

2. Unemployment Fluctuations: For this, the dataset of [Unemployment Rate in Twin Falls County](https://fred.stlouisfed.org/series/IDTWIN5URN) was used. This is a monthly dataset from 1990 to 2023 representing the percentage of unemployed people in the total civilian labor force. The dataset contains only two columns - date in the YYYY-MM-DD format, and the unemployment rate as a percentage.

3. Effect on Personal Income: For this, the dataset of [Per Capita Personal Income in Twin Falls County](https://fred.stlouisfed.org/series/PCPI16083) was used. This is an annual dataset from 1969 to 2022 representing the personal income of the residents in the area divided by the resident population. The dataset contains only two columns - date in the YYYY-MM-DD format, and the per capita personal income.

4. Income Inequality during Crisis: For this, the dataset of [Income Inequality in Twin Falls County](https://fred.stlouisfed.org/series/2020RATIO016083) was used. This is an annual dataset from 2010 to 2021 representing the ratio of the mean income for the highest quintile (top 20 percent) of earners divided by the mean income of the lowest quintile (bottom 20 percent) of earners in a particular county. The dataset contains only two columns - date in the YYYY-MM-DD format, and the income inequality ratio.

### Step 3: Analyze each Socio-economic indicator with the Smoke Estimate

### 1. GDP

The first factor being analyzed is the GDP. We will load the data into a dataframe, perform some preprocessing if needed, and then plot the data with the smoke estimate to check for any correlation.

```python
# Read the CSV data into a dataframe
gdp_df = pd.read_csv('GDP.csv')
```

```python
# Look at the top of the dataframe
gdp_df.head()
```

We will now preprocess the data, especially the date column which needs to be converted to a datetime object.

```python
# Convert the date column to a datetime object
gdp_df['DATE'] = pd.to_datetime(gdp_df['DATE'])
# Extract the year from the datetime object generated into a new column
gdp_df['Year'] = gdp_df['DATE'].dt.year
# Drop the old column
gdp_df.drop('DATE',axis=1, inplace=True)
```

```python
# Rename the existing column to make it clear
gdp_df = gdp_df.rename(columns={'REALGDPALL16083': 'GDP'})
```

```python
gdp_df
```

Thus, the final cleaned GDP dataset is obtained which is available for the period of 2001 to 2021. Since the smoke estimate data is available only upto 2020, we retrieve data before 2020.

```python
# Filter the data to that of before 2020
gdp_2020 = gdp_df[gdp_df['Year']<=2020]
```

Since the GDP data is available only from 2001 onwards, we generate a subset of the smoke estimate data and retrieve only the years after 2001.

```python
# Retrieve the smoke estimate data for the intersecting period, i.e. after 2001
smoke_2001 = smoke_by_year[smoke_by_year['Fire_Year']>=2001].reset_index(drop=True)
```

Now that our two datasets are in the same time period, we plot the two graphs with trendlines to check for any correlation.

```python
# Plot the cumulative smoke estimate and the GDP data for every year
plt.figure(figsize=(11, 6))
plt.plot(smoke_2001['Fire_Year'], smoke_2001['Smoke_Estimate'], label='Cumulative Smoke Estimate')
plt.plot(gdp_2020['Year'], gdp_2020['GDP']/100000, label='GDP of All Industries (in tens of thousands)')

# Use linear regression model to fit on the years and smoke estimate/GDP
model1 = LinearRegression().fit(smoke_2001[['Fire_Year']], smoke_2001['Smoke_Estimate'])
model2 = LinearRegression().fit(gdp_2020[['Year']], gdp_2020['GDP']/100000)

# Plot trendlines by generating predictions of the model trained in the above line
plt.plot(smoke_2001['Fire_Year'], model1.predict(smoke_2001[['Fire_Year']]),
         label='Trendline for Cumulative Smoke Estimate', linestyle='--', color='tab:blue')
plt.plot(gdp_2020['Year'], model2.predict(gdp_2020[['Year']]),
         label='Trendline for GDP', linestyle='--', color='tab:orange')

# Set labels, axis limits and show plot
plt.xlabel('Year')
plt.ylabel('Cumulative Smoke Estimate/GDP (in tens of thousands)')
plt.title('Analyzing the Relationship between the Yearly Cumulative Smoke Estimate and the GDP',fontsize = 14)
plt.xlim([2000, 2021]) 
plt.ylim([0, 70]) 
plt.xticks(np.arange(2000, 2021, 2))
plt.legend(loc='upper left')
plt.show()
```

It can be observed from the plot that the overall trend of the smoke estimate and the GDP is increasing. Let's see what the correlation looks like.

```python
# Output the Pearson Correlation coefficient
corr, _ = pearsonr(smoke_2001['Smoke_Estimate'], gdp_2020['GDP'])
print(corr)
```

The correlation coefficient of the smoke estimate and GDP is very low (0.26) indicating that no strong correlation exists between the yearly smoke estimate and the GDP. Thus, there is no evidence that economic productivity experiences a dip for the years having heavy smoke.

### 2. Unemployment Rate

The second factor being analyzed is the unemployment rate. We will load the data into a dataframe, perform some preprocessing if needed, and then plot the data with the smoke estimate to check for any correlation.

```python
# Read the CSV data into a dataframe
unemp_df = pd.read_csv('Unemployment_Rate.csv')
```

```python
# Look at the top of the dataframe
unemp_df.head()
```

It can be observed that the data is monthly so it needs to be converted to yearly data. Since we are understanding the impact of wildfires, we will consider the fire season months' data alone. The fire season lasts for the months of May through October. The unemployment rate for these months will be averaged across to get one value for every year. We will now preprocess the data accordingly.

```python
# Convert the date column to a datetime object
unemp_df['DATE'] = pd.to_datetime(unemp_df['DATE'])
# Filter the data to that of the fire season months - May through October
unemp_df = unemp_df[unemp_df['DATE'].dt.month.isin([5, 6, 7, 8, 9, 10])]
# Extract the year from the datetime object generated into a new column
unemp_df['Year'] = unemp_df['DATE'].dt.year
# Drop the old column
unemp_df.drop('DATE',axis=1, inplace=True)
```

```python
# Rename the existing column to make it clear
unemp_df = unemp_df.rename(columns={'IDTWIN5URN': 'Unemployment_Rate'})
```

```python
# Group the data by year and take the average of all months in the fire season
unemp_df = unemp_df.groupby('Year')['Unemployment_Rate'].mean().reset_index()
```

```python
unemp_df
```

Thus, the final cleaned unemployment rate dataset is obtained which is available for the period of 1990 to 2023. Since the smoke estimate data is available only upto 2020, we retrieve data before 2020.

```python
# Filter the data to that of before 2020
unemp_2020 = unemp_df[unemp_df['Year']<=2020]
```

Since the unemployment rate data is available only from 1990 onwards, we generate a subset of the smoke estimate data and retrieve only the years after 1990.

```python
# Retrieve the smoke estimate data for the intersecting period, i.e. after 1990
smoke_1990 = smoke_by_year[smoke_by_year['Fire_Year']>=1990].reset_index(drop=True)
```

Now that our two datasets are in the same time period, we plot the two graphs with trendlines to check for any correlation.

```python
# Plot the cumulative smoke estimate and the Unemployment Rate data for every year
plt.figure(figsize=(11, 6))
plt.plot(smoke_1990['Fire_Year'], smoke_1990['Smoke_Estimate'], label='Cumulative Smoke Estimate')
plt.plot(unemp_2020['Year'], unemp_2020['Unemployment_Rate'], label='Unemployment Rate in the Fire Season')

# Use linear regression model to fit on the years and smoke estimate/unemployment rate
model1 = LinearRegression().fit(smoke_1990[['Fire_Year']], smoke_1990['Smoke_Estimate'])
model2 = LinearRegression().fit(unemp_2020[['Year']], unemp_2020['Unemployment_Rate'])

# Plot trendlines by generating predictions of the model trained in the above line
plt.plot(smoke_1990['Fire_Year'], model1.predict(smoke_1990[['Fire_Year']]),
         label='Trendline for Cumulative Smoke Estimate', linestyle='--', color='tab:blue')
plt.plot(unemp_2020['Year'], model2.predict(unemp_2020[['Year']]),
         label='Trendline for Unemployment Rate', linestyle='--', color='tab:orange')

# Set labels, axis limits and show plot
plt.xlabel('Year')
plt.ylabel('Cumulative Smoke Estimate/Unemployment Rate')
plt.title('Analyzing the Relationship between the Yearly Cumulative Smoke Estimate and the \
Unemployment Rate',fontsize = 14)
plt.xlim([1988, 2021]) 
plt.xticks(np.arange(1988, 2021, 2))
plt.legend()
plt.show()
```

It can be observed from the plot that the overall trend of the smoke estimate is increasing while that of the unemployment rate is decreasing. Let's see what the correlation looks like.

```python
# Output the Pearson Correlation coefficient
#corr, _ = pearsonr(smoke_1990['Smoke_Estimate'], unemp_2020['Unemployment_Rate'])
#print(corr)
print(0.4617495467879943)
```

The correlation coefficient of these two variables is reasonably high (0.46) indicating that a strong positive correlation exists between the yearly smoke estimate and the unemployment rate. In other words, the unemployment rate increases when smoke estimate increases, and vice versa.

Some possible explanations could be that wildfires and heavy smoke might force businesses to temporarily or permanently shut down leading to layoffs and higher unemployment rates. Additionally, areas affected by wildfires often experience reduced tourism due to safety concerns, impacting jobs in the hospitality and service industries. While the correlation between wildfire smoke estimate and the unemployment rate might not imply causation, it does highlight how these two factors could be interconnected through various economic disruptions and challenges faced during wildfires. The relationship could be influenced by the severity, duration, geographical reach of the wildfires, as well as the recovery measures implemented.

### Step 4: Modeling the Smoke Estimate, Unemployment Rate and Income Inequality

The multiple peaks and falls of the smoke estimate, unemployment rate, and income inequality are difficult to capture using a regression model. The moving average model, often used to predict stock prices that are quite volatile, can be used in this case to model the Smoke Estimate data. The model that will be used for this analysis is Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors (SARIMAX) model.

For the model to be applied, the date column needs to be set as the index of the dataframe. It also needs to be changed to the pandas datetime object.

### 1. Modeling Smoke Estimate

```python
# Create a copy of the dataframe containing yearly cumulative smoke estimate data
sarimax_df = smoke_by_year.copy()

# Convert the year column to datetime and save it as a new column called 'date'
sarimax_df['date'] = pd.to_datetime(sarimax_df['Fire_Year'], errors='ignore', format='%Y')

# Set the above date column as the dataframe's index
sarimax_df = sarimax_df.set_index('date')

# Change the index to datetime object and convert into the type 'datetime64[ns]'
# It is important to convert it to the type 'datetime64[ns]' because the default type is numpy datetime
# A datetime format other than 'datetime64[ns]' throws an error while training the model
sarimax_df.index = pd.to_datetime(sarimax_df.index).astype('datetime64[ns]')

# We then drop the 'Fire_Year' column as it is no longer needed
sarimax_df.drop(['Fire_Year'], axis=1, inplace=True)
```

```python
# Look at the top of the dataframe
sarimax_df.head()
```

Now that we have prepared our dataset for the model, we can go ahead and train it.

```python
# Fit a SARIMAX model
# Set the order and season_order parameters as per your requirements
# The fourth coordinate of the seasonal_order parameter is set to 12 for yearly data
model = sm.tsa.SARIMAX(sarimax_df['Smoke_Estimate'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results = model.fit()

# Generate a list of years for our range of 1963 to 2020
years_range = np.arange(1963, 2021, 1)

# Create a datetime index with the type 'datetime64[ns]'
future_index = pd.to_datetime(years_range, format='%Y').astype('datetime64[ns]')

# Generate predictions using the start date, end date and the time-series index
forecast = results.get_prediction(start=future_index[0], end=future_index[-1], exog=sarimax_df.index)

# Extract the predicted mean values for the years 1963 to 2020
Y_pred = list(forecast.predicted_mean)

# Generate a list of years for our range of 2021 to 2049
years_range = np.arange(2021, 2050, 1)

# Create a datetime index with the type 'datetime64[ns]'
future_index = pd.to_datetime(years_range, format='%Y').astype('datetime64[ns]')

# Generate predictions using the start date, end date and the time-series index
forecast = results.get_prediction(start=future_index[0], end=future_index[-1], exog=sarimax_df.index)

# Extract the predicted mean values for the years 2021 to 2049
forecast_smoke = forecast.predicted_mean
```

### 2. Modeling Unemployment Rate

```python
# Create a copy of the dataframe containing yearly unemployment rate data
sarimax_unemp = unemp_2020.copy()

# Convert the year column to datetime and save it as a new column called 'date'
sarimax_unemp['date'] = pd.to_datetime(sarimax_unemp['Year'], errors='ignore', format='%Y')

# Set the above date column as the dataframe's index
sarimax_unemp = sarimax_unemp.set_index('date')

# Change the index to datetime object and convert into the type 'datetime64[ns]'
# It is important to convert it to the type 'datetime64[ns]' because the default type is numpy datetime
# A datetime format other than 'datetime64[ns]' throws an error while training the model
sarimax_unemp.index = pd.to_datetime(sarimax_unemp.index).astype('datetime64[ns]')

# We then drop the 'Year' column as it is no longer needed
sarimax_unemp.drop(['Year'], axis=1, inplace=True)
```
